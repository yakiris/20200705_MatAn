# Тема "Понятие о производной"

### 1. Найти производную выражения:

### a. $sinx * cosx$  
$y' = (sinx)' * cosx + sinx * (cosx)' = cos^2x - sin^2x = cos(2x)$

```python
plt.grid()
x = np.linspace(-2, 2, 100)
y = [np.sin(i)*np.cos(i) for i in x]
y2 = [np.cos(2*i) for i in x]
plt.plot(x, y, label='y')
plt.plot(x, y2, label='y\'')
plt.legend()
```

### b. $ln(2x + 1)^3$  
$y = 3ln(2x + 1)$  
$y' = 3(ln(2x + 1))' = 3 \frac{(2x + 1)'}{2x + 1} = \frac{6}{2x + 1}$

```python
plt.grid()
x = np.linspace(-0.5, 1, 100)
y = [np.log10(2*i + 1)**3 for i in x]
y2 = [6/(2*i + 1) for i in x]
plt.plot(x, y, label='y')
plt.plot(x, y2, label='y\'')
plt.legend()
```

### c. $\sqrt{sin^2(ln(x^3))}$

$y' = |sin(3 ln x)| = \frac{sin(ln(x^3))}{|sin(ln(x^3))|} * cos(3 ln x) * \frac{3}{x}$

### d. $\frac{x^4}{ln(x)}$
$y' = \frac{(x^4)'ln(x) - x^4(ln(x))'}{ln()^2} = \frac{x^3(4ln(x)-1)}{2ln(x)}$

```python
plt.grid()
x = np.linspace(-10, 10, 50)
y = [(i**4/(np.log10(i))) for i in x]
y2 = [((i**3)*(4*np.log10(i)-1))/(2*np.log10(i)) for i in x]
plt.plot(x, y, label='y')
plt.plot(x, y2, label='y\'')
plt.legend()
```

### 2. Найти выражение производной функции и ее значение в точке:

### $f(x) = cos(x^2 + 3x), x_0 = \sqrt{\pi}$

$f'(x) = -sin(x^2 + 3x) * (2x + 3)$  
$f(\sqrt \pi) = -sin(\pi + 3\sqrt \pi) * (2 \sqrt \pi + 3) \simeq -0.96$

```python
plt.grid()
x = np.linspace(3, 4, 100)
y = [(np.cos(i**2 + 3*i)) for i in x]
plt.plot(x, y, label='y')
plt.legend()
```

### 4. Найти угол наклона касательной к графику функции в точке:

### $f(x) = \sqrt{3x} * lnx, x_0 = 1$

$f'(x) = (\sqrt 3x)' * lnx + (lnx)' * \sqrt 3x = \frac{3lnx}{2x} + \frac{\sqrt 3x}{x}$  
$f(x_0) = \sqrt 3 = tg\alpha$  
$\alpha = 60^0$

```python
plt.grid()
x = np.linspace(-5, 5, 100)
y = [(np.sqrt(3*i)*np.log10(i)) for i in x]
plt.plot(x, y, label='y')
plt.legend()
```
