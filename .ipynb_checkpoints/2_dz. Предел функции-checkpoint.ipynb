```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

# Тема “Предел функции”

### 1. Предложить пример функции, не имеющей предела в нуле и в бесконечностях.

$$f(x) = sin(\frac{x + 1}{x})$$

```python
plt.grid()
x = np.linspace(-10, 10, 51)
y = [np.sin((i + 1)/i) for i in x]
plt.plot(x, y)
```

### 2. Привести пример функции, не имеющей предела в точке, но определенной в ней.

$$f(x) = sign(x)$$

```python
plt.grid()
x = np.linspace(-10, 10, 51)
y = [np.sign(i) for i in x]
plt.plot(x, y)
```

### 3. Исследовать функцию $f(x) = x^3 - x^2$ по плану:

**a. Область задания и область значений.**  $x \in(- \infty; + \infty), y \in (- \infty; + \infty)$

**b. Нули функции и их кратность.**  $x_1 = 0 (кратность 2); x_2 = 1(кратность 1)$

**c. Отрезки знакопостоянства.**  - $(- \infty; 0] [0; + \infty)$

**d. Интервалы монотонности.**  - возрастает: $(- \infty; 0) U (\sim 0.7; + \infty)$; убывает: $(0; \sim 0.7)$

**e. Четность функции.**  - функция общего вида

**f. Ограниченность.** - неограниченная 

**g. Периодичность.** - не периодическая

```python
plt.grid()
x = np.linspace(-1, 1.5, 50)
y = [(i**3 - i**2) for i in x]
plt.plot(x, y)
```

### 4. Найти предел:

### a. $\lim\limits_{x \rightarrow 0} \frac{3x^3 - 2x^2}{4x^2} = \frac{3x - 2}{4} = - \frac{1}{2}$ 

```python
plt.grid()
x = np.linspace(-10, 10, 50)
y = [((3*i - 2)/4) for i in x]
plt.plot(x, y)
```

# Тема “Теоремы о пределах”

### 1. Найти предел:

### a. $\lim\limits_{x \rightarrow 0} \frac{sin(2x)}{4x} = \frac{1}{2}$

```python
plt.grid()
x = np.linspace(-10, 10, 50)
y = [(np.sin(2*i)/4*i) for i in x]
plt.plot(x, y)
```

### b. $\lim\limits_{x \rightarrow 0} \frac{x}{sin(x)} = 1$

```python
plt.grid()
x = np.linspace(-10, 10, 50)
y = [(i/np.sin(i)) for i in x]
plt.plot(x, y)
```

### c. $\lim\limits_{x \rightarrow 0} \frac{x}{arcsin(x)} = 1$

```python
plt.grid()
x = np.linspace(-10, 10, 50)
y = [(i/np.arcsin(i)) for i in x]
plt.plot(x, y)
```

### d. $\lim\limits_{x \rightarrow \infty} (\frac{4x + 3}{4x - 3})^{6x} = \lim\limits_{x \rightarrow \infty} (\frac{4x - 3 + 3 + 3}{4x - 3})^{6x} = \lim\limits_{x \rightarrow \infty} (1 + \frac{6}{4x - 3})^{\frac{4x - 3}{6} * \frac{6}{4x - 3} * 6x} =  e^9$

```python
plt.grid()
x = np.linspace(-10, 10, 20)
y = [((4*i + 3)/(4*i - 3))**(6*i) for i in x]
plt.plot(x, y)
```
